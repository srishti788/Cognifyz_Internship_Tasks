```python


import pandas as pd 
df = pd.read_csv("Dataset .csv")# Total number of restaurants
total_restaurants = len(df)

# Percentage offering table booking
table_booking_percent = df['Has Table booking'].value_counts(normalize=True) * 100
print("Table Booking (%):")
print(table_booking_percent)

# Percentage offering online delivery
online_delivery_percent = df['Has Online delivery'].value_counts(normalize=True) * 100
print("\nOnline Delivery (%):")
print(online_delivery_percent)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Table Booking Pie Chart
df['Has Table booking'].value_counts().plot.pie(autopct='%1.1f%%', title='Table Booking Availability', colors=['lightgreen', 'salmon'])
plt.ylabel('')
plt.show()

# Online Delivery Pie Chart
df['Has Online delivery'].value_counts().plot.pie(autopct='%1.1f%%', title='Online Delivery Availability', colors=['skyblue', 'orange'])
plt.ylabel('')
plt.show()
```

```python
# Group by 'Has Table booking' and calculate mean of 'Aggregate rating'
avg_rating_by_table_booking = df.groupby('Has Table booking')['Aggregate rating'].mean()

print("Average Ratings Based on Table Booking Availability:")
print(avg_rating_by_table_booking)
```

```python
# Group by 'Price range' and 'Has Online delivery'
online_delivery_by_price = df.groupby(['Price range', 'Has Online delivery']).size().unstack()

print("Online Delivery Availability by Price Range:")
print(online_delivery_by_price)
```

```python
online_delivery_percent = online_delivery_by_price.div(online_delivery_by_price.sum(axis=1), axis=0) * 100

print("\nPercentage of Online Delivery by Price Range:")
print(online_delivery_percent.round(2))
```
