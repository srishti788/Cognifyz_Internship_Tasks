```python
import pandas as pd
df = pd.read_csv("Dataset .csv")


# Find the most common price range
most_common_price_range = df['Price range'].value_counts().idxmax()
count = df['Price range'].value_counts().max()

print(f"The most common price range is: {most_common_price_range} (Count: {count})")
```

```python
price_dist = df['Price range'].value_counts().sort_index()
print("Price Range Distribution:")
print(price_dist)
```

```python
import matplotlib.pyplot as plt

price_dist.plot(kind='bar', color='skyblue', figsize=(7, 5))
plt.title('Distribution of Restaurants by Price Range')
plt.xlabel('Price Range (1 = Low, 4 = High)')
plt.ylabel('Number of Restaurants')
plt.xticks(rotation=0)
plt.show()
```

```python
# Calculate average rating for each price range
avg_rating_by_price = df.groupby('Price range')['Aggregate rating'].mean().sort_index()

print("Average Rating for Each Price Range:")
print(avg_rating_by_price.round(2))
```

```python
import matplotlib.pyplot as plt

avg_rating_by_price.plot(kind='bar', color='coral', figsize=(7, 5))
plt.title('Average Restaurant Rating by Price Range')
plt.xlabel('Price Range (1 = Low, 4 = High)')
plt.ylabel('Average Aggregate Rating')
plt.ylim(0, 5)
plt.xticks(rotation=0)
plt.show()
```

```python
# Group by 'Price range' and find average rating & most common rating color
rating_summary = df.groupby('Price range').agg({
    'Aggregate rating': 'mean',
    'Rating color': lambda x: x.mode()[0]  # most common color in that price range
}).sort_values(by='Aggregate rating', ascending=False)

print("Average Rating and Rating Color by Price Range:")
print(rating_summary)
```
