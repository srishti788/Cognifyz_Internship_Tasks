```python

print(df['Country Code'].value_counts())


# df['Country'] = df['Country Code'].map(country_map)

# Plot
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Country Code', order=df['Country Code'].value_counts().index)
plt.title('Distribution of Country Code')
plt.xlabel('Country Code')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
print(df['City'].value_counts().head(10))  # Top 10 cities

# Plot top 10 cities
top_cities = df['City'].value_counts().nlargest(10).index

plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['City'].isin(top_cities)], x='City', order=top_cities)
plt.title('Top 10 Cities with Most Records')
plt.xlabel('City')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Top 10 most frequent cuisines
print(df['Cuisines'].value_counts().head(10))

# Plot top 10
top_cuisines = df['Cuisines'].value_counts().nlargest(10).index

plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['Cuisines'].isin(top_cuisines)], x='Cuisines', order=top_cuisines)
plt.title('Top 10 Cuisines')
plt.xlabel('Cuisines')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.show()
```

```python
from collections import Counter

# Split and flatten cuisines
cuisine_series = df['Cuisines'].dropna().apply(lambda x: [c.strip() for c in x.split(',')])
cuisine_list = [item for sublist in cuisine_series for item in sublist]

# Count
cuisine_counts = pd.Series(Counter(cuisine_list)).sort_values(ascending=False)
print(cuisine_counts.head(10))

# Plot
cuisine_counts.head(10).plot(kind='bar', figsize=(10, 5), color='orange')
plt.title('Most Common Individual Cuisines')
plt.xlabel('Cuisine')
plt.ylabel('Count')
plt.show()
```

```python
from collections import Counter

# Droping NaNs and split cuisines by comma
cuisine_series = df['Cuisines'].dropna().apply(lambda x: [c.strip() for c in x.split(',')])
cuisine_list = [item for sublist in cuisine_series for item in sublist]

# Count individual cuisines
cuisine_counts = pd.Series(Counter(cuisine_list)).sort_values(ascending=False)
print("Top Individual Cuisines:")
print(cuisine_counts.head(10))
```
