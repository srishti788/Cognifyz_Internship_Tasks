```python
# Convert 'Aggregate rating' to numeric (if it's not already)
df['Aggregate rating'] = pd.to_numeric(df['Aggregate rating'], errors='coerce')

# Convert date columns if any
# df['Date'] = pd.to_datetime(df['Date'])
```

```python
# Count unique values
print(df['Aggregate rating'].value_counts().sort_index())

# Plot distribution
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='Aggregate rating', palette='viridis')
plt.title('Distribution of Aggregate Rating')
plt.xlabel('Aggregate Rating')
plt.ylabel('Count')
plt.show()
```

```python
class_counts = df['Aggregate rating'].value_counts(normalize=True)
print("Class distribution ratio:\n", class_counts)
```
